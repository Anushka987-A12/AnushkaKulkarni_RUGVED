```python
matches=pd.read_csv(r"C:\Users\anush\Downloads\matches.csv")
print("q4")
df=matches.groupby(['toss_winner','toss_decision']).count().filter(['id']).rename(columns={'id':''})
df.index.set_names(['team','toss descision'],level=[0,1],inplace=True)
df
```

```python
print("q7")
print("teams that won with highest number of runs")
df=matches.groupby(["toss_winner","win_by_runs"]).sum().sort_values('win_by_runs',ascending=False).filter(['win_by_runs']).head(10)
df.index.set_names(['team','win by runs'],level=[0,1],inplace=True)
df
```

```python
print("q7")
print("teams that won with lowest number of runs")
df=matches.groupby(["toss_winner","win_by_runs"]).sum().sort_values('win_by_runs').filter(['win_by_runs']).head(10)
df.index.set_names(['team','win by runs'],level=[0,1],inplace=True)
df
```
